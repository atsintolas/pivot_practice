```python
import pandas as pd
import numpy as np

df = pd.read_excel('AlgebraGradesData.xlsx')
df.head()
```

```python
df.loc[df['Grade']=='A', 'passingGrade'] = 1
df.loc[df['Grade']=='B', 'passingGrade'] = 1
df.loc[df['Grade']=='C', 'passingGrade'] = 1
df.loc[df['Grade']=='D', 'passingGrade'] = 0
df.loc[df['Grade']=='F', 'passingGrade'] = 0
df.head()
```

```python
#%age of students with a passing grade
pd.pivot_table(df, index = 'passingGrade', values=['Fname'], aggfunc='count')

#can then determine percentage
```

```python
#%age of women with a passing grade
pd.pivot_table(df, index = ['Gender','passingGrade'], values=['Fname'], aggfunc='count')

#can then do percentage
```

```python
#average hours of study for all students
df['Hours of Study'].mean()
```

```python
#average hours of study for students with passing grade
pd.pivot_table(df, index ='passingGrade', values=['Hours of Study'], aggfunc='mean')
```

```python
#max hours of study for students that have a D
pd.pivot_table(df, index ='Grade', values=['Hours of Study'], aggfunc='max')
```

```python
#avg hours for men with C
pd.pivot_table(df, index ='Grade', columns = 'Gender', values=['Hours of Study'], aggfunc='mean')
```
